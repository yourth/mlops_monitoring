# Relational Data Hierarchical Modeling

>
> [SCV.HMA1](https://sdv.dev/SDV/user_guides/relational/hma1.html)
>

```python
from sdv import load_demo
```

```python
metadata, tables = load_demo(metadata=True)
metadata
```

```python
metadata.visualize()
```

```python
tables
```

```python
from sdv.relational import HMA1

import warnings
warnings.filterwarnings('ignore')

model = HMA1(metadata)
model.fit(tables)
```

```python
# get sample data
new_data = model.sample()
new_data
```

```python
# save hierarchical model
model.save('./models/my_model.pkl')
```

```python
# get 5 rows of top level sample data
model.sample(num_rows=5)
```

```python
# get 5 rows from subset of tables
model.sample('sessions', num_rows=5)
```

```python
# get pandas dataframe of sample data
model.sample('users', num_rows=5, sample_children=False)
```
